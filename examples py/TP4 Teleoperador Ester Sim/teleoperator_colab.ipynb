# Creamos funciones de comunicación
Ytilizaremos está función para gestionar la comunicación por API

```python
def requestPost(id,key,title,comando,valor):
    robot_name = "Robot" + str(id)
    data={"apikey": "acc$2g1","user_id": "Grupo UAI","id":id,"title":title,"key":key,"command":comando,"value":valor}
    response = requests.put('http://localhost:8000/Ester', json=data)

    print("Status code: ", response.status_code)
    print("Printing Entire Post Request")
    print(response.json())
```
